```python
# 파이썬의 파일 쓰기 기능을 이용하기
# 파일 쓰기 기능을 이용하여
# 로또 번호 중복없는 6개의 번호로 차례로 구성된
# 결과를 10가지로 추천하여
# "로또추천.txt" 라는 파일로 남기라.

import random
import os

# 파일 저장 위치 확인용 (선택사항)
print("파일 저장 위치:", os.getcwd())

with open("로또추천.txt", "w", encoding="utf-8") as f:
    f.write("로또 추천 번호 10세트\n\n")

    for game in range(1, 11):
        # 1~45 중 중복없는 6개 자동 생성
        lotto = random.sample(range(1, 46), 6)

        lotto.sort()
        line = ", ".join(f"{n:02d}" for n in lotto)
        f.write(f"{game:2}번째 : {line}\n")

print("완료! -> 로또추천.txt 생성됨")
```

```python
# 불편한 코드
result1 = 0
result2 = 0

def add1(num):
    global result1
    result1 += num  # 결괏값(result)에 입력값(num) 더하기
    return result1  # 결괏값 리턴

def add2(num):
    global result2
    result2 += num  # 결괏값(result)에 입력값(num) 더하기
    return result2  # 결괏값 리턴  
```

```python
# 클래스 사용

class Cal:
    def __init__(self):
        self.result = 0
        
    def add(self, num):
        self.result += num
        return self.result

    def mul(self, num):
        self.result *= num
        return self.result
```

```python
cal1 = Cal()
cal2 = Cal()
```

```python
print(cal1.add(3))
```

```python
print(cal1.mul(3))
```

```python
# 예제1

class DDuk:
    def __init__(self, won, ea):
        self.won = won
        self.ea = ea

    def dduk(self):
        result = self.won * self.ea
        print(f"배가고파서 떡볶이집에서 떡볶이를 {result}원어치 먹었다.")

    def odeng(self):
        result = self.won * self.ea
        print(f"배가고파서 떡볶이집에서 오뎅을 {result}원어치 먹었다.")
```

```python
재형 = DDuk(2500,4)
호성 = DDuk(1500,7)
재훈 = DDuk(2000,5)
```

```python
재형.dduk()
재형.odeng()
```

```python
호성.dduk()
호성.odeng()
```

```python
재훈.dduk()
재훈.odeng()
```

```python
# 예제2
class Pet:
    def __init__(self, name, species):
        """펫을 생성"""
        self.name = name
        self.species = species
        self.hunger = 50 # 배고픔 수치 0 ~ 100
    def feed(self, many = 10):
        """밥을 주면 배고픔 감소"""
        if self.hunger > 10:
            self.hunger -= many
        return f"{self.name}에게 밥을 줬어요. 배고픔:{self.hunger}"
    def play(self, how = 20):
        """놀아주면 배고픔 증가"""
        if self.hunger < 80:
            self.hunger += how
        return f"{self.name}와 잘 놀았어요~!! 배고픔:{self.hunger}"
```

```python
dog1 = Pet("뽀삐","말티즈")
dog2 = Pet("강쥐","허숙히")
cat1 = Pet("컁이","도둑고양이")
cat2 = Pet("킁이","집고양이")
```

```python
cat1.feed(20)
```

```python
cat1.play(30)
```

```python
print(cat1.feed())
print(cat1.play())
```
