```python
# 배열과 튜블을 각각 name 과 age, grade 변수에 담는 내용(list comprehension으로 해결하는 코드)
students = [("jane",22,"A"),("dave",32,"B"),("sally",17,"B")]

grade = 0
[name, age, grade] = [[s[i] for s in students] for i, x in enumerate([name, age, grade])]
print(name, age, grade, sep ='\n')
```

```python
# 제너레이터로 파일속의 에러내용을 반환하는 로직(yield)
def error_lines(filepath):
    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:
            if "ERROR" in line:
                yield line.strip()

# 사용 예시
for err in error_lines("app.log"):
    print("에러 발견:",err)
```
