```python
# 기본코드 (공청기 필터1)
test_air = [1,23,32,-5,45,0,-3,123,2,-12]
def pos(num):
    result = []
    for n in num:
        if n > 0:
            result.append(n)
    return result

print(pos(test_air))
```

```python
# 기본코드응용 (리스트컴프리헨션)
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print([n for n in test_air if n > 0])
```

```python
# filter ** 중요 filter(함수, 반복_가능한_데이터)
# 공청기2용 필터 기본코드 // 필터는 리턴이 True !! 가 중요!

test_air = [1,23,32,-5,45,0,-3,123,2,-12]
def pos_fn(n):
    return n > 0
def neg_fn(n):
    return n < 0
print(list(filter(pos_fn ,test_air)))
print(list(filter(neg_fn ,test_air)))
```

```python
# filter 응용
test_air = [1,23,32,-5,45,0,-3,123,2,-13]
print(list(filter(lambda x: x > 0 ,test_air)))
print(list(filter(lambda x: x < 0 ,test_air)))
```

```python
# 기본코드(가공 1)
sample = [1, 2, 3, 4, 11, 22, 33, 44]
def two_times(numberlist):
    result = []
    for number in numberlist:
        result.append(number*2)
    return result

result = two_times(sample)
print(result)
```

```python
# 1. 클래스 상속받고 메서드 추가하기1
class Calculator:
    def __init__(self, value=0):
        self.value = value
        
    def add(self, val):
        self.value += val
        return self.value

    def minus(self, val):
        self.value -= val
        return self.value


class UpgradeCalculator(Calculator):
    def multiply(self, val):
        self.value *= val
        return self.value


cal = UpgradeCalculator()

cal.add(10)    
cal.minus(7)  

print(cal.value)
```

```python
# 2. 클래스 상속받고 메서드 추가하기2
cal.value < 120

class Calculator:
    def __init__(self, value=0):
        self.value = value
        
    def add(self, val):
        self.value += val
        return self.value

    def minus(self, val):
        self.value -= val
        return self.value

class MaxLimitCalculator(Calculator):
    def add(self, val):
        self.value += val
        if self.value > 100:
            self.value = 100 
        return self.value

cal = MaxLimitCalculator()

cal.add(50)
cal.add(70)

print(cal.value)
```
