```python
def counter_maker():
    count = [0]  # 리스트로 감싸서 mutable하게 만듦

    def counter():
        count[0] += 1
        return count[0]

    return counter

c1 = counter_maker()
c2 = counter_maker()
```

```python
c1()
```

```python
c2()
```

```python
print(c1.__closure__[0].cell_contents)
print(c2.__closure__[0].cell_contents)
```

## 데코레이터


```python
import time

def myfunc():
    start = time.time()
    print("함수가 실행됩니다.")
    print("함수 수행시간: %f 초" % (time.time()-start))
myfunc()
```

```python
import time

def outfunc(X):
    def infunc():
        start = time.time()
        print("함수 준비.")
        result = X()        
        print("함수 수행시간: %f 초" % (time.time()-start))
        return result
    return infunc
    
@outfunc
def Xfunc():
    print("함수가 실행됩니다.")
@outfunc   
def Yfunc():
    print("함수가 어렵습니다.")
```

```python
Xfunc()
Yfunc()
```

```python
Xfunc = outfunc(Xfunc)
Yfunc = outfunc(Yfunc)
Xfunc()
Yfunc()
```

```python
# decorator.py
import time

def elapsed(original_func):   # 기존 함수를 인수로 받는다.
    def wrapper():
        start = time.time()
        result = original_func()    # 기존 함수를 수행한다.
        end = time.time()
        print("함수 수행시간: %f 초" % (end - start))  # 기존 함수의 수행시간을 출력한다.
        return result  # 기존 함수의 수행 결과를 반환한다.
    return wrapper

@elapsed
def myfunc():
    print("함수가 실행됩니다.")

# decorated_myfunc = elapsed(myfunc)  # @elapsed 데코레이터로 인해 더이상 필요하지 않다.
# decorated_myfunc()

myfunc()
```

```python
# 이터레이터만들기
class MyIterator:
    def __init__(self, data):
        self.data = data
        self.position = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.position >= len(self.data):
            raise StopIteration
        result = self.data[self.position]
        self.position += 1
        return result
```

```python
it = MyIterator([1,2,3])
```

```python
for n in it:
    print(n)
```

```python
# 제너레이터
def mybot():
    print("작업시작")
    yield "1번작업완료"
    print("로봇A 위치로 동작")
    yield "2번작업완료"
    print("로봇A 그리퍼 동작후 위치반환")
    yield "3번작업완료"
    print("로봇B 가동시작")    
    yield "end"        
gb = mybot()
```

```python
try:
    next(gb)
except StopIteration as e:
    print("더이상 할 작업이 없습니다.", e)
```
