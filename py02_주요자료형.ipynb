```python
from fastapi import FastAPI
import requests

app = FastAPI()

def get_price():
    btc = requests.get(
        "https://api4.binance.com/api/v3/ticker/price?symbol=BTCUSDT"
    ).json()["price"]

    rate = requests.get(
        "https://www.koreaexim.go.kr/site/program/financial/exchangeJSON",
        params={"authkey":"5NI1wgZX5OHJXox7MUewT3TwrjmvStFA","data":"AP01"}
    ).json()

    usd = [i for i in rate if i["cur_unit"]=="USD"][0]["deal_bas_r"]

    return float(btc) * float(usd.replace(",",""))

@app.get("/btc")
def btc():
    return {"krw": get_price()}
```
